```python
#Check for common errors
import polars as pl

# 1. Force Polars to never shrink or truncate text in the console
pl.Config.set_fmt_str_lengths(5000)  # Allow strings up to 5000 chars
pl.Config.set_tbl_cols(-1)           # Show all columns
pl.Config.set_tbl_width_chars(500)   # Maximize console width

def extract_and_display_errors(file_paths):
    for model_name, path in file_paths.items():
        print("\n" + "="*80)
        print(f" ERROR ANALYSIS: {model_name.upper()} (Top 10 cases with 0% Citation Recall)")
        print("="*80)
        
        try:
            df = pl.read_csv(path)
            
            # Filter for rows where the model failed to get the exact citation
            error_df = df.filter(pl.col("cit_recall") == 0.0)
            
            # Handle the column name difference (you used 'answer' in toys, 'raw_model_response' in full)
            pred_col = "answer" if "answer" in df.columns else "raw_model_response"
            
            # Select only the columns you requested (plus ID so you can track them)
            cols_to_show = ["id", "prompt", "correct_answer", "sources", pred_col, "cit_recall", "bleu", "rouge_l", "bertscore_f1"]
            
            # Ensure we only select columns that actually exist in the CSV to prevent errors
            available_cols = [c for c in cols_to_show if c in error_df.columns]
            
            # Grab the top 10
            top_10_errors = error_df.select(available_cols).head(10)
            
            # 2. Print them block-by-block for easy human reading
            for i, row in enumerate(top_10_errors.iter_rows(named=True), 1):
                print(f"\n--- {model_name} | ERROR CASE {i} | ID: {row.get('id', 'N/A')} ---")
                print(f"PROMPT (Legal Case) : {row.get('prompt', '')}")
                print(f"REQUIRED CITATION   : {row.get('sources', '')}")
                print(f"CORRECT ANSWER      : {row.get('correct_answer', '')}")
                print(f"MODEL GENERATED     : {row.get(pred_col, '')}")
                print(f"SCORES              : BERTScore: {row.get('bertscore_f1', 0):.2%} | ROUGE-L: {row.get('rouge_l', 0):.2%} | BLEU: {row.get('bleu', 0):.2%}")
                
        except Exception as e:
            print(f"Failed to process {path}: {e}")
   
```

```python
files_to_analyze = {
    "Claude 4.5 Haiku": "/mnt/red/red_hanka_bcthesis/llm/final_evaluated_claude_results.csv",
    "FineTuned RoBERTa": "/mnt/red/red_hanka_bcthesis/llm/final_evaluated_roberta_results.csv",
    "RoBERTa + Mistral RAG": "/mnt/red/red_hanka_bcthesis/llm/final_evaluated_rag_results.csv"
}
extract_and_display_errors(files_to_analyze)
```
